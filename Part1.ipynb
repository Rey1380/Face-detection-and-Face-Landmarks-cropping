```python
import cv2
face_detector=cv2.CascadeClassifier(r'C:\Users\Lenovo\Desktop\haarcascade_frontalface_default.xml')
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('C:\\Users\\Lenovo\\Desktop\\faces.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("image")
plt.show()
n=0
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
results = face_detector.detectMultiScale(gray,1.15,3)
for (x,y,w,h) in results:
    if(y!=374):
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
        n=n+1;
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("final image")
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
print("number of faces=",n)
```

```python
import cv2
import sys
import time
faceCascade = cv2.CascadeClassifier(r'C:\Users\Lenovo\Desktop\haarcascade_frontalface_default.xml')
video_capture = cv2.VideoCapture(0)
max1=0
while True:

#Calculating FPS
    # Capture frame-by-frame
    # displaying the frame with fps
    ret, frame = video_capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30),
        flags=cv2.CASCADE_SCALE_IMAGE
    )

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
    # Display the resulting frame
    cv2.imshow('Video', frame)

#finding FPS by using CV_CAP_PROP_FPS
    (major_ver, minor_ver, subminor_ver) = (cv2.__version__).split('.')
    if int(major_ver)  < 3 :
        fps = video_capture.get(cv2.cv.CV_CAP_PROP_FPS)
        print("Frames per second using video.get(cv2.cv.CV_CAP_PROP_FPS): {0}".format(fps))
    else :
        fps = video_capture.get(cv2.CAP_PROP_FPS)
        print("Frames per second using video.get(cv2.CAP_PROP_FPS) : {0}".format(fps))

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    num_frames = 8;
    print("Capturing {0} frames".format(num_frames))
    # Start time
    start = time.time()
    # Grab a few frames
    for i in range(0, num_frames) :
        ret1, frame1 = video_capture.read()

    # End time
    end = time.time()
    # Time elapsed
    seconds = end - start
    print ("Time taken : {0} seconds".format(seconds))
    # Calculate frames per second
    fps  = num_frames / seconds
    print("Estimated frames per second : {0}".format(fps))
    if(fps>max1):
        max1=fps



print("FPS MAX:",max1)
# When everything is done, release the capture
video_capture.release()
cv2.destroyAllWindows()
```
